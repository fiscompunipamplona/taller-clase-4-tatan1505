```python
from pylab import plot, show

y = [1., 2.4, 1.7, 0.3, 0.6, 1.8]

plot(y)
show()
```

```python
x = [0.5, 1., 2., 4., 7., 10.]

plot(x,y)
show()
```

```python
from pylab import savefig
plot(x, y)
savefig('destination_path.eps', format='eps', dpi=1000)
show()
```

```python
from numpy import linspace, sin, cos

x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid
x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
ylim( -1.1, 1.1)
xlim(0., 10)
grid("on")
show()
```

```python
from pylab import xlabel, ylabel
x = linspace(0, 10, 1000)
y = sin(x)
plot(x,y)
ylim( -1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje x")
ylabel("Eje y")
show()
```

```python
x = linspace(0, 10, 100)
y = sin(x)

plot(x,y,'ko')
show()
```

```python
y = sin(x)
plot(x,y,'k-')
y = cos(x)
plot(x,y,'k--')
grid("on")
show()
```
